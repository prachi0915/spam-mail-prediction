```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
# data collection and preprocessing
mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
mail_data.head()
```

```python
#repalce null value
maildata = mail_data.where((pd.notnull(mail_data)),'')
```

```python
maildata.head()
```

```python
maildata['Category'].value_counts()
```

```python
#label encodeing
maildata.loc[maildata['Category']=='spam','Category',]=0
maildata.loc[maildata['Category']=='ham','Category',]=1
```

```python
maildata.isnull().sum()
```

```python
X = maildata['Message']
Y = maildata['Category']
```

```python
maildata['Category'].value_counts()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
```

```python

# Feature Extraction
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
feature_extraction.fit(X_train)

X_train_features = feature_extraction.transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(Y)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
X_train_prediction = model.predict(X_train_features)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test_features)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(testing_data_accuracy)
```

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]
input_data_features = feature_extraction.transform(input_mail)
prediction = model.predict(input_data_features)
print(prediction)
if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
